```python
import seaborn as sns
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset('tips')
tips.info()
```

```python
import matplotlib.pyplot as plt
sns.lineplot(data=tips,x='total_bill',y='tip')
plt.title("Bill v/s Tips")
plt.grid(True)
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
sns.barplot(data=tips,x='total_bill',y='tip')
```

```python
sns.boxplot(data=tips,x='total_bill',y='tip')
```

```python
flights = sns.load_dataset('flights')
flights.info()
```

```python
flights
```

```python
pv = flights.pivot(index='month',columns='year',values='passengers')
pv
```

```python
sns.heatmap(pv, annot=True, fmt='d', cmap='YlGnBu')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
planets = sns.load_dataset('planets')
print(planets.info())

plt.figure(figsize=(10, 6))
sns.countplot(data=planets, x='method',color='orange')
plt.xticks(rotation=45)
plt.title("Planets Discovered by Method")
plt.show()
```
