```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y1 = [i*2 for i in x]
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.grid(True)
plt.xlabel("Numbers")
plt.ylabel("Double")
plt.title("Double of numbers")

y2 = [i**2 for i in x]
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.grid(True)
plt.xlabel("Numbers")
plt.ylabel("Square")
plt.title("Square the number")
plt.tight_layout()
plt.show()
```

```python
y3 = [i ** 0.5 for i in x]
y4 = [10 - i for i in x]

plt.figure(figsize=(8, 6))
plt.suptitle('Comparision of numbers')

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('x * 2')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('x squared')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('sqrt(x)')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('10 - x')

plt.tight_layout()
plt.show()
```
