```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[4,5,6])
```

```python
years = [2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
celeb = ['yes','yes','yes','yes','yes','yes','yes','no','no','no','no','no','no','no']
plt.plot(years,celeb,'v:g')
plt.xlabel('YEARS',)
plt.ylabel('isCELEB')
plt.title('unknown birth year celeb.')
plt.show()
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
plt.plot(years,kohli,'o-r',label='Kohli runs')
plt.plot(years,sehwag,'o-b',label='Sehwag runs')
plt.xlabel('Years')
plt.ylabel('Runs scores')
plt.legend()
plt.grid(True)
plt.title('Performance comparison')
```

```python
import numpy as np

for i in range(50):
    plt.plot(np.random.rand(100), linewidth=1)

plt.title("Too Much Data Can Be Confusing!")
plt.grid(True)
plt.tight_layout()
plt.show()
```
