```python
import matplotlib.pyplot as plt
study_hours = [1,2,3,4,5,6,7,8]
exam_score = [32,40,95,63,72,36,80,93]
colors = ['red' if score<50 else 'green' for score in exam_score]
size = 70
sizes = [size if score>=50 else size*0.3 for score in exam_score]
plt.scatter(study_hours,exam_score,c=colors,s=sizes)
plt.xlabel('Hours')
plt.ylabel('Score')
plt.title('Exam preparation Hrs v/s score')
plt.grid(True)
```

```python
import matplotlib.pyplot as plt
study_hours = [1,2,3,4,5,6,7,8]
exam_score = [32,40,95,63,72,36,80,93]
plt.scatter(study_hours,exam_score,c=exam_score,cmap='plasma')
plt.colorbar(label='Score')
plt.xlabel('Hours')
plt.ylabel('Score')
plt.title('Exam preparation Hrs v/s score')
plt.grid(True)
```
