```python
import matplotlib.pyplot as plt
labels=['Bhanu','Rajesh','Anil','Jaswanth','Ravi']
marks=[60,90,97,70,75]
colors=['#F05979','#F059CD','#6B59F0','#59F0AA','#DEF059']
explode=[0,0,0.5,0,0]
plt.pie(
    marks,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor':'black','linewidth':2,'linestyle':'--'},
    explode=explode
)
plt.show()
```
